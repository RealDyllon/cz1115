# Exercise 2 - Problem 2

## (a) Find the summary statistics of SalePrice from the numeric dataframe from Problem 1
Summary statistics are: mean, median, quartiles, etc.

```python
salesprice_stats = train_data_numeric['SalePrice'].describe()
print(salesprice_stats)
```

## (b) Visualise the summary statistics and distribution of SalePrice
Using Box-Plot, Histogram, KDE.

```python
# BoxPlot
f = plt.figure(figsize=(10, 10))
sb.boxplot(data = salesprice_stats, orient = "h")
```

```python
# Histogram
f = plt.figure(figsize=(10, 10))
sb.histplot(data = salesprice_stats)
```

```python
# KDE
f = plt.figure(figsize=(10, 10))
sb.kdeplot(data = salesprice_stats)
```

## (c) Find the Summary Statistics of LotArea using the numeric dataframe

```python
lotarea_stats = train_data_numeric['LotArea'].describe()
print(lotarea_stats)
```

## (d) Visualise the summary statistics and distribution of LotArea using standard Box-Plot, Histogram, KDE

```python
# Box Plot
f = plt.figure(figsize=(10,10))
sb.boxplot(data = lotarea_stats, orient = "h")
```

```python
# Histogram
f = plt.figure(figsize=(10, 10))
sb.histplot(data = lotarea_stats)
```

```python
# KDE
f = plt.figure(figsize=(10, 10))
sb.kdeplot(data = lotarea_stats)
```

## (e) Plot SalePrice vs Lot-Area using jointplot, then find the correlation between the two

```python
sb.jointplot(data = train_data_numeric, x = "LotArea", y = "SalePrice", height = 10)
```
